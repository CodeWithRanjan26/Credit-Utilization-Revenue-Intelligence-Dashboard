```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("UCI_Credit_Card.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df["Total_Spend"] = df[[
    "BILL_AMT1","BILL_AMT2","BILL_AMT3",
    "BILL_AMT4","BILL_AMT5","BILL_AMT6"
]].sum(axis=1)
```

```python
df[["Total_Spend"]].head()
```

```python
df["Utilization_%"] = df["Total_Spend"] / df["LIMIT_BAL"]
```

```python
df[["Utilization_%"]].head()
```

```python
def segment(u):
    if u < 0.3:
        return "Low"
    elif u <= 0.7:
        return "Medium"
    else:
        return "High"
    
df["Utilization_Segment"] = df["Utilization_%"].apply(segment)
```

```python
df[["Utilization_%","Utilization_Segment"]].head()
```

```python
df["Total_Payment"] = df[[
    "PAY_AMT1","PAY_AMT2","PAY_AMT3",
    "PAY_AMT4","PAY_AMT5","PAY_AMT6"
]].sum(axis=1)
```

```python
df[["Total_Payment"]].head()
```

```python
df["Risk_Flag"] = df["default.payment.next.month"]
```

```python
df[["Risk_Flag"]].head()
```

```python
df["Avg_Delay"] = df[[
    "PAY_0","PAY_2","PAY_3",
    "PAY_4","PAY_5","PAY_6"
]].mean(axis=1)

df[["Avg_Delay"]].head()
```

```python
df.head()
```

```python
df.to_csv("final_credit_data.csv", index=False)
```
